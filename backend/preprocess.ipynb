```python
import pandas as pd
import numpy as np
```

```python
#download from "https://www.reinfolib.mlit.go.jp/realEstatePrices/""
df = pd.read_csv('../raw_data/aichi_20071_20234.csv', encoding="cp932", on_bad_lines='skip')
```

```python
df = df.drop(columns=['市区町村コード'])
df = df.drop(columns=['価格情報区分'])
df = df.drop(columns=['建物の構造'])
df = df.drop(columns=['用途'])
df = df.drop(columns=['今後の利用目的'])
df = df.drop(columns=['都市計画'])
df = df.drop(columns=['建ぺい率（％）'])
df = df.drop(columns=['容積率（％）'])
df = df.drop(columns=['改装'])
df = df.drop(columns=['取引の事情等'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['取引価格（総額）'] = df['取引価格（総額）']/10000
```

```python
df['取引時期（年）'] = df['取引時期'].str[:4]
```

```python
# Get the current year
current_year = pd.Timestamp.now().year
```

```python
# Step 1: Remove '年' character
df['建築年'] = df['建築年'].str.replace('年', '')

# Step 2: Drop rows with the value '戦前'
df = df[df['建築年'] != '戦前']

# Step 3: Convert to numeric, coercing errors to NaN
df['建築年'] = pd.to_numeric(df['建築年'], errors='coerce')

# Step 4: Fill NaN values with a placeholder or drop them
df['建築年'].fillna(0, inplace=True)  # Example: Fill with 0, or you can choose to drop these rows

# Get the current year
current_year = pd.Timestamp.now().year

# Step 5: Calculate the age of the building and add as new column '築年数'
df['築年数'] = current_year - df['建築年'].astype(int)
```

```python
df['取引時期（年）'] = df['取引時期（年）'].astype(int)
df['建築年'] = df['建築年'].astype(int)
```
